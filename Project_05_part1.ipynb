## Importing & merging many files (easy case)

3. __Load__ all files (yob????.txt) and __merge/concatenate__ all files into one Pandas DataFrame. Make sure you add the __column "Year"__.

4. Create a __RangeIndex__ and __save__ the DataFrame (with columns "Year", "Name", "Gender" and "Count") in a new csv-file. 

```python
import pathlib

YoB_DF = pd.read_csv('yob1880.txt', sep=',', names=['Name', 'Gender', 'Count'])
YoB_DF['Year'] = 1880
year = 1881
while True:
    if pathlib.Path(f'yob{year}.txt').exists():
        df = pd.read_csv(f'yob{year}.txt', sep=',', names=['Name', 'Gender', 'Count'])
        df['Year'] = year
        year += 1
        YoB_DF = pd.concat([YoB_DF, df])
    else:
        break

YoB_DF
```

```python
YoB_DF.to_csv('YoB_full.csv')
pd.read_csv('YoB_full.csv')
```
