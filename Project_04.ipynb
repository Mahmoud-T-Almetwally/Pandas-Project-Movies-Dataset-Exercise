## Creating an SQLite Database

1. __Import__ sqlite3 (as sq3) and __create__ a new SQLite Database with the name __"movies.db"__.

```python
import sqlite3 as sq3
import pandas as pd

con = sq3.connect('movies.db')
```

## Loading Data from DataFrames into an SQLite Database

2. __Load__ the json file __"some_movies.json"__ and __split__ the dataset into the following __four datasets__ (save each dataset in a Pandas DataFrame).

__Dataset #1 (Movies)__ with columns ["id", "title", "revenue", "budget", "belongs_to_collection_name", "release_date"]. <br>
Convert "release_date" to datetime and transform "budget" and "revenue" to Million USD before loading into the Database. 

__Dataset #2 (Votes)__ with columns ["id", "vote_count", "vote_average"]. 

__Dataset #3 (Genres)__ with columns ["genre_id", "genre_name", "id"]. <br> 

__Dataset #4 (Prod)__ with columns ["comp_id", "comp_logo_path", "comp_name", "comp_origin_country", "id" ]. <br>


```python
df = pd.read_json('some_movies.json')
df.head(5)
```

```python
df.info(verbose=0)
```

```python
import ast
import numpy as np

df.belongs_to_collection.apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else np.nan)
df.belongs_to_collection
```

```python
df.belongs_to_collection = df.belongs_to_collection.apply(lambda x: x['name'] if isinstance(x, dict) else np.nan)
df.belongs_to_collection
```

```python
df.release_date = pd.to_datetime(df.release_date)
df.release_date
```

```python
Movies = pd.DataFrame({'id':df.id,
                       'title':df.title,
                       'revenue_musd':df.revenue.div(1_000_000),
                       'budget_musd':df.budget.div(1_000_000),
                       'belongs_to_collection':df.belongs_to_collection,
                       'release_date':df.release_date})
```

```python
Movies.info()
```

```python
Votes = pd.DataFrame({'id':df.id,
                      'vote_count':df.vote_count,
                      'vote_average':df.vote_average})

Votes
```

```python
df.genres[0]
```

```python
Genres =  df[['genres', 'id']].explode(column='genres')
Genres['genre_id'] = Genres.genres.apply(lambda x: x['id'])
Genres['genre_name'] = Genres.genres.apply(lambda x: x['name'])
Genres.drop(columns='genres', inplace=True)
```

```python
Genres
```

```python
Prod = df[['production_companies', 'id']].explode('production_companies')
```

```python
Prod['production_companies'][0]
```

```python
Prod['comp_id'] = Prod['production_companies'].apply(lambda x: x['id'])
Prod['comp_logo_path'] = Prod['production_companies'].apply(lambda x: x['logo_path'])
Prod['comp_origin_country'] = Prod['production_companies'].apply(lambda x: x['origin_country'])
Prod['comp_name'] = Prod['production_companies'].apply(lambda x: x['name'])
Prod.drop(columns='production_companies', inplace=True)
```

```python
Prod
```

```python
Prod.comp_origin_country.value_counts()
```

```python
Prod.replace('', np.nan, inplace=True)
```

3. __Load__ the datasets __into the database__ (each dataset should be a separate table in the database). __Name__ the tables "Movies", "Votes", "Genres", "Prod".

```python
Prod.to_sql('Prod', con=con, index=False, if_exists='replace')
```

```python
Movies.to_sql('Movies', con=con, index=False, if_exists='replace')
```

```python
Genres.to_sql('Genres', con=con, index=False, if_exists='replace')
```

```python
Votes.to_sql('Votes', con=con, index=False, if_exists='replace')
```

## Join Queries

7. __Perform__ the following __SQL Join Queries__ and __store__ the results in DataFrames:

__Join "Movies" and "Votes"__ (all columns).

```python
movies = pd.read_sql_query('SELECT * FROM Movies INNER JOIN Votes ON Movies.id = Votes.id', con=con)
movies
```

__Join "Movies" and "Votes" (only the columns "id", "title", "vote_average").__

```python
movies = pd.read_sql_query('SELECT Movies.id, Movies.title, Votes.vote_average FROM Movies INNER JOIN Votes ON Movies.id = Votes.id', con=con)
movies
```

__Join "Movies" and "Votes" (only the columns "id", "title", "vote_average") and return only those movies with vote_average > 8.__

```python
movies = pd.read_sql_query('SELECT Movies.id, Movies.title, Votes.vote_average FROM Movies INNER JOIN Votes ON Movies.id = Votes.id WHERE Votes.vote_average > 8', con=con)
movies
```

__Join "Movies" and "Votes" (only the columns "id", "title", "vote_average") and return only those movies with vote_average > 8 and in ascending budget order__.

```python
movies = pd.read_sql_query('SELECT Movies.id, Movies.title, Votes.vote_average FROM Movies INNER JOIN Votes ON Movies.id = Votes.id WHERE Votes.vote_average > 8 ORDER BY Movies.budget_musd ', con=con)
movies
```

## Final Case Study

8. __Perform__ the following advanced __SQL Queries__ and __store__ the results in DataFrames:

__Get the Total Revenue (sum) for each Production Company.__

```python
movies = pd.read_sql_query('SELECT sum(Movies.revenue_musd), Prod.comp_name FROM Movies INNER JOIN Prod ON Movies.id = Prod.id GROUP BY Prod.comp_id', con=con)
movies
```

__Get all Production Companies for the movie "Titanic".__

```python
comps = pd.read_sql_query('SELECT Prod.comp_name FROM Prod INNER JOIN Movies ON Movies.id = Prod.id WHERE Movies.title = "Titanic"', con=con)
comps
```

__Get the Total Revenue (sum) for each Genre.__

```python
revs = pd.read_sql_query('SELECT sum(Movies.revenue_musd), Genres.genre_name FROM Movies INNER JOIN Genres ON Movies.id = Genres.id GROUP BY Genres.genre_id', con=con)
revs
```

__Get all Genres for the movie "Frozen II".__

```python
genres = pd.read_sql_query('SELECT Genres.genre_name FROM Genres INNER JOIN Movies ON Movies.id = Genres.id WHERE Movies.title = "Frozen II"', con=con)
genres
```
